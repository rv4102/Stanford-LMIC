```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
from joblib import dump, load
```

```python
train_path = '/kaggle/input/mchmlmic-new/train.parquet.gzip'
test_path = '/kaggle/input/mchmlmic-new/test.parquet.gzip'
pred_cols = ["Mean_BMI","Median_BMI","Unmet_Need_Rate","Under5_Mortality_Rate","Skilled_Birth_Attendant_Rate","Stunted_Rate"]
```

```python
train = pd.read_parquet(train_path)
train
```

```python
X = train.drop(pred_cols, axis=1)
y = train[pred_cols]
```

```python
def mcrmse(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred, multioutput='raw_values')
    rmse = np.sqrt(mse)
    return np.average(rmse)

model = RandomForestRegressor(n_estimators=800, max_features=0.5, bootstrap=True, oob_score=mcrmse, random_state=42, n_jobs=-1, verbose=2)
model.fit(X, y)
```

```python
model.oob_score_
```

```python
dump(model, '/kaggle/working/rf.joblib')
```

```python
test = pd.read_parquet(test_path)

y_pred = model.predict(test)
out = pd.DataFrame(y_pred, columns=pred_cols)
out['DHSID'] = test.index
out = out[['DHSID'] + pred_cols]
out.to_csv('/kaggle/working/sub.csv', index=False)
```
