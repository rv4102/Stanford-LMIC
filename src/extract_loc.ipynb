```python
import dask.dataframe as dd
```

```python
def convert_new_ind(x):
    x_new = x[:2] + x[3:7] + x[8:]
    return x_new
df = dd.read_csv('../Data/gee_features.csv', sample=100000000, converters={'new_ind': convert_new_ind},
                       usecols=lambda x: x in ['DHSID','DHSREGNA','DHSYEAR','LATNUM','LONGNUM',
                                                   'SOURCE','URBAN_RURA'], dtype={
    'DHSID': 'object',
    'DHSREGNA': 'object',
    'DHSYEAR': 'object',
    'LATNUM': 'float64',
    'LONGNUM': 'float64',
    'SOURCE': 'object',
    'URBAN_RURA': 'object'
})
```
