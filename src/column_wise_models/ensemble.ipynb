```python
from joblib import dump, load
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
pred_cols = ["Mean_BMI","Median_BMI","Unmet_Need_Rate","Under5_Mortality_Rate","Skilled_Birth_Attendant_Rate","Stunted_Rate"]
```

```python
nulls = pd.read_csv('data/nulls.csv')
nulls = nulls[nulls['0'] > 0.0]['Unnamed: 0'].values
```

```python
low_imp_features = load('data/low_imp_features.joblib')
```

```python
df = pd.read_parquet('data/main_data')
df = df[~df.index.duplicated(keep='first')]
df.drop(columns=nulls, inplace=True)
df
```

```python
other_df = pd.read_parquet('data/other_data/')
other_df = other_df[~other_df.index.duplicated(keep='first')]
other_df
```

```python
transformer = make_column_transformer(
    (OneHotEncoder(), ['DHSCC', 'URBAN_RURA', 'DHSREGNA']),
    remainder='passthrough')

transformed = transformer.fit_transform(other_df)
transformed_df = pd.DataFrame(transformed.todense(), columns=transformer.get_feature_names_out())
transformed_df.index = other_df.index
transformed_df
```

```python
df.drop(columns=[col for col in df.columns if col in low_imp_features], inplace=True)
df
```

```python
transformed_df.drop(columns=[col for col in transformed_df.columns if col in low_imp_features], inplace=True)
transformed_df
```

```python
df = df.join(transformed_df, how='left')
df
```

```python
labels = pd.read_csv('data/training_label.csv', index_col='DHSID')
labels.drop(columns=['URBAN_RURA', 'DHSYEAR', 'DHSCLUST', 'LATNUM', 'LONGNUM'], inplace=True)
labels
```

```python
df = df.join(labels, how='left')
df
```

```python
df.to_parquet('data/top_features_data.parquet.gzip')
```

```python
sub = pd.read_csv('data/sample submission.csv', index_col='DHSID')
train = df[~df.index.isin(sub.index)]
train
```

```python
df.drop(columns=pred_cols, inplace=True)
test = sub.join(df, how='left')
test.drop(columns=pred_cols, inplace=True)
test
```

```python
pred_cols
```

```python
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.7)]
```

```python
X = df.drop(to_drop, axis=1)
test_X = test[X.columns]
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
```

```python
import gc
sub = pd.read_csv('best_submission.csv')

for i, pcol in enumerate(pred_cols):
    print(f'[STARTED] ({i+1}/{len(pred_cols)}) LABEL: {pcol}')
    y = labels[pcol]
    y = y[~y.isna()]
    
    X_new = pd.merge(X, y, on='DHSID', how='inner')
    
    X = X_new.drop(pcol, axis = 1)
    y = X_new[pcol]
    
    hgbr = HistGradientBoostingRegressor(max_iter=2000, max_depth=20)
    hgbr.fit(X, y)
    
    sub[pcol] = sub[pcol] * 0.6 + hgbr.predict(test_X) * 0.4
    sub.to_csv('submission.csv', index=False)
    print(f'[COMPLETED] ({i+1}/{len(pred_cols)}) LABEL: {pcol}')
    
    gc.collect()
```
